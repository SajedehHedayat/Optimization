# GD

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import time
```

```python
np.random.seed(42)
x = np.random.rand(100000,1)
y = 2*x**2 + .1*np.random.randn(100000,1)
```

```python
plt.plot(x,y, 'o')
plt.title('Generated Data (train)')
plt.grid()
plt.xlabel('x_train')
plt.ylabel('y_train')
```

```python
# initializes parameters "a" and "b" randomly

np.random.seed(42)
b = np.random.randn(1)

# print('initial value of b : ', b)
# sets learning rate
lr = 0.5
# defines number of epochs
n_epochs = 50
start = time.time()
for epoch in range(n_epochs):
  # computes our model's predicted output

  y_hat = b * x**2
  #print(y_hat)
  error = (y - y_hat)
  # computes MSE
  loss = (error ** 2).mean()
  # computes gradients for "b" parameters
  b_grad = -2 * (x**2 * error).mean()
  # updates parameter using gradients and the learning rate
  b = b - lr * b_grad
end = time.time()
print("final value of b : ", b)
print(end - start)
```

# SGD

```python
np.random.seed(42)
x = np.random.rand(100000,1)
y = 2*x**2 + .1*np.random.randn(100000,1)
```

```python
plt.plot(x,y, 'o')
plt.title('Generated Data (train)')
plt.grid()
plt.xlabel('x_train')
plt.ylabel('y_train')
```

```python
# initializes parameters "a" and "b" randomly
np.random.seed(42)

b = np.random.randn(1)
batch_size = 4
# print('initial value of b : ', b)
# sets learning rate
lr = 0.5
# defines number of epochs
n_epochs = 50
start = time.time()

for epoch in range(n_epochs):
  # computes our model's predicted output
  indexes = np.random.randint(0, len(x), batch_size) # random sample

  Xs = np.take(x, indexes)
  ys = np.take(y, indexes)
  y_hat = b * Xs**2
  #print(y_hat)
  error = (ys - y_hat)
  # computes MSE
  loss = (error ** 2).mean()
  # computes gradients for both "a" and "b" parameters
  b_grad = -2 * (Xs**2 * error).mean()
  # updates parameters using gradients and the learning rate
  b = b - lr * b_grad
end = time.time()
print("final value of b : ", b)
print('time : ', (end - start))
```
